```python
import os
import  args_parse
import torch
from torchvision import transforms

import datasets
import datasets.utils as du
import net.models as models
```

```python
ROOT = os.path.dirname(os.path.dirname(os.path.abspath('__file__')))

MODELS = [
    'vae_439'
]

DATASETS = [
    'gravityspy'
]

FLAGS = args_parse.parse_common_options(
    x_dim=439,
    batch_size=4,
    num_epochs=100,
    num_workers=4,
    log_steps=10,
    lr=1e-3,
    model={
        'choices': MODELS,
        'default': 'vae_439',
    },
    dataset={
        'choices': DATASETS,
        'default': 'gravityspy',
    }
)

fake_data_len = 100000
loader = du.SampleGenerator(
    data=(torch.randn(FLAGS.batch_size, 1, FLAGS.x_dim, FLAGS.x_dim),
                torch.randint(low=0, high=9, size=(FLAGS.batch_size, 1), dtype=torch.int64)),
    sample_count=fake_data_len // FLAGS.batch_size
)

model = getattr(models, FLAGS.model)()
optimizer = torch.optim.Adam(model.parameters(), lr=FLAGS.lr)

model.train()
for step, (data, target) in enumerate(loader):
    params = model(data, target, return_params=True)
    loss = model.criterion(**params)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(loss)
    
    
```
